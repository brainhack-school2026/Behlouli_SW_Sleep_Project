```python
# on charge les librairies 
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# on check les chemins 
PROJECT_ROOT = Path("..")

RAW_FILE = PROJECT_ROOT / "data" / "raw" / "raw.xlsx"

print(RAW_FILE)
print("File exists:", RAW_FILE.exists())
```

```python
# on charge 
df = pd.read_excel(RAW_FILE)

df.head()
```

```python
# on vérifie les dimensions 
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
# les colonnes 
df.columns.tolist()
```

```python
# les fichiers et les participants 
print("Unique filenames:", df["filename"].nunique())
print("Unique participants:", df["ID participant"].nunique())
```

```python
# les électrodes conservées vs exclu
keep_electrodes = [
    "F3", "F4",
    "C3", "C4",
    "P3", "P4",
    "O1", "O2"
]

excluded_electrodes = [
    e for e in sorted(df["chan_label"].dropna().unique())
    if e not in keep_electrodes
]

print("Kept electrodes:", keep_electrodes)
print("Excluded electrodes:", excluded_electrodes)
```
